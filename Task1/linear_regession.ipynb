Implementation of linear regression using Scikit-learn library

```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# Load the insurance dataset
insurance_data = pd.read_csv('insurance.csv')

# Prepare the feature matrix X and target variable y
X = insurance_data[['age', 'bmi']]
y = insurance_data['charges']

# Normalize the feature matrix X
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_norm, y)

# Predict charges
y_predict = model.predict(X_norm)

# Plotting the results
plt.scatter(y, y_predict, color='blue', label='Actual')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', label='predicted')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Actual Charges vs Predicted Charges')
plt.show()
```
